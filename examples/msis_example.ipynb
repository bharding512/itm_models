# Run an altitude profile

```python
ds = itm_models.msis(
    time=pd.to_datetime('2025-07-24 12:59'),
    lat=0,
    lon=0,
    alt=np.linspace(80,300,100),
)
```

```python
import matplotlib.pyplot as plt

plt.figure()
ds.O.plot(xscale='log', y='alt')
ds.N2.plot(xscale='log', y='alt')
```

# Run over lat/lon

```python
ds = itm_models.msis(
    time=pd.to_datetime('2025-07-24 12:59'),
    lat=np.linspace(0,90),
    lon=np.linspace(0,360),
    alt=250.
)
```

```python
plt.figure()
ds.TEMPERATURE.plot()
```
